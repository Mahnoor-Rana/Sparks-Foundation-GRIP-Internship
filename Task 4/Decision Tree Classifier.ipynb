## The Aim of this project is to create Decision Tree Classifier and visualize it graphically.
### The purpose is to feed the new data to this classifier , it would be able to predict the right class accordingly.

### IMPORTING LIBRARIES

```python
import pandas as pd 
import sklearn.datasets as datasets
```

### LOADING THE DATASET

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris.data , columns=iris.feature_names)
df.head()
```

```python
y = iris.target
y
```

### Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(df,y)
```

### Visualize the Decision Tree

```python
!pip install pydotplus
!apt-get install graphviz -y
```

```python
from six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
```

### Visualize the graph

```python
dot_data = StringIO()
export_graphviz(tree,out_file = dot_data,feature_names = iris.feature_names, filled = True, 
                rounded = True, special_characters = True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```
